```python
df_complete_wo = df_complete_wo.drop(columns=['Anæstesi melder klar'])
df_complete_wo = df_complete_wo.drop(columns=['Anæstesistart'])
df_complete_wo = df_complete_wo.drop(columns=['Anæstesistop'])
df_complete_wo = df_complete_wo.drop(columns=['Forsinkelsesårsag'])
```

```python
df_complete_wo = df_complete_wo.drop(columns=['Aktionsdiagnose - Kode & tekst'])
df_complete_wo = df_complete_wo.drop(columns=['I opvågning'])
```

```python
datetime_variables = ['Dato', 'Pt ankommet til hospitalet', 'Planlagt stue klargøring start', 'Planlagt stue klargøring start', 'Stue klargøring start', 'Stue klargjort', 'Patient på stuen', 'Patient på stuen (Planlagt)', 'Procedure start', 'Procedure slut', 'Patient klar til afgang', 'Patient forlader stuen (Planlagt)', 'Patient forlader stuen', 'Stue rengjort (Planlagt)', 'Stue rengøring start', 'Stue rengjort', 'Patient forlader afdeling']
```

```python
# Beregn procentdelen af NaN i kolonne
nan_procent = df_complete_wo['Klar til udskrivelse efter opvågning'].isna().mean() * 100

print(f"Procentdel af manglende værdier i 'Klar til udskrivelse efter opvågning': {nan_procent:.2f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Gruppér data efter 'Speciale' og beregn antal og gennemsnitlig forsinkelse
stats_df = df_complete_wo.groupby('Speciale')['Forsinkelse (minutter)'].agg(['count', 'mean']).reset_index()

# Omdøb kolonnerne for overskuelighed
stats_df.columns = ['Speciale', 'Antal_Operationer', 'Gns_Forsinkelse']

# 2. Filtrér specialer med meget få operationer (valgfrit - her sat til min. 5 operationer)
# Dette gøres for at undgå at specialer med f.eks. kun 1 operation skævvrider billedet
stats_df = stats_df[stats_df['Antal_Operationer'] >= 10]

# 3. Beregn korrelationen (Pearson)
korrelation = stats_df['Antal_Operationer'].corr(stats_df['Gns_Forsinkelse'])

print(f"Antal specialer medtaget i analysen: {len(stats_df)}")
print(f"Korrelation mellem antal operationer og gennemsnitlig forsinkelse: {korrelation:.2f}")

# 4. scatter plot til visualisering
plt.figure(figsize=(12, 7))
plt.scatter(stats_df['Antal_Operationer'], stats_df['Gns_Forsinkelse'], 
            alpha=0.6, s=100, color='teal', edgecolor='black')

# Tilføj navne på de enkelte specialer ved hvert punkt (hvis der ikke er for mange)
for i, row in stats_df.iterrows():
    plt.annotate(row['Speciale'], 
                 (row['Antal_Operationer'], row['Gns_Forsinkelse']),
                 xytext=(5, 5), textcoords='offset points', fontsize=9)

# Tilføj titler og labels
plt.title('Sammenhæng mellem Volumen og Gennemsnitlig Forsinkelse pr. Speciale', fontsize=14)
plt.xlabel('Antal Operationer (Volumen)', fontsize=12)
plt.ylabel('Gennemsnitlig Forsinkelse (minutter)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Gruppér efter 'Speciale'
# Vi bruger .agg() til at lave navngivne kolonner med det samme (Named Aggregation)
stats_df = df_complete_wo.groupby('Speciale').agg(
    Gns_Alder=('Patient Alder', 'mean'),
    Gns_Forsinkelse=('Forsinkelse (minutter)', 'mean'),
    Antal_Operationer=('Case-ID Anonymous', 'count')
).reset_index()

# 2. Filtrér specialer med færre end 10 operationer (valgfrit - du kan ændre tallet)
# Dette sikrer, at gennemsnittet er baseret på et solidt grundlag
stats_df = stats_df[stats_df['Antal_Operationer'] >= 10]

# 3. Beregn korrelationen mellem alder og forsinkelse
korrelation = stats_df['Gns_Alder'].corr(stats_df['Gns_Forsinkelse'])

print(f"Antal specialer med mindst 10 operationer: {len(stats_df)}")
print(f"Korrelation mellem gennemsnitsalder og forsinkelse: {korrelation:.2f}")

# 4. Lav visualiseringen
plt.figure(figsize=(12, 7))
plt.scatter(stats_df['Gns_Alder'], stats_df['Gns_Forsinkelse'], 
            alpha=0.6, s=150, color='teal', edgecolor='black')

# Tilføj navne på specialerne ved hvert punkt
for i, row in stats_df.iterrows():
    plt.annotate(row['Speciale'], 
                 (row['Gns_Alder'], row['Gns_Forsinkelse']),
                 xytext=(5, 5), textcoords='offset points', fontsize=9)

# Tilføj titler og labels
plt.title('Sammenhæng mellem Patienternes Gennemsnitsalder og Forsinkelse pr. Speciale', fontsize=14)
plt.xlabel('Gennemsnitsalder (år)', fontsize=12)
plt.ylabel('Gennemsnitlig Forsinkelse (minutter)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Udtræk 'Time på dagen' og 'Ugedag'
df_complete_wo['Time'] = df_complete_wo['Procedure start'].dt.hour
df_complete_wo['Ugedag'] = df_complete_wo['Procedure start'].dt.day_name()

# 3. Sorter ugedage så de ikke står alfabetisk
dage_orden = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_complete_wo['Ugedag'] = pd.Categorical(df_complete_wo['Ugedag'], categories=dage_orden, ordered=True)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Sørg for at tidsformatet er korrekt
df_complete_wo['Procedure start'] = pd.to_datetime(df_complete_wo['Procedure start'])

# 2. Udtræk timen
df_complete_wo['Time'] = df_complete_wo['Procedure start'].dt.hour

# 3. Lav lineplottet opdelt på Akut case (J/N)
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=df_complete_wo, 
    x='Time', 
    y='Forsinkelse (minutter)', 
    hue='Akut case (J/N)',   
    marker='o', 
    estimator='mean', 
    errorbar=None            
)


plt.axhline(0, color='red', linestyle='--', alpha=0.5)

# Formatering af grafen
plt.title('Gennemsnitlig forsinkelse: Akutte vs. Planlagte operationer')
plt.xlabel('Tidspunkt på dagen (Time)')
plt.ylabel('Forsinkelse (minutter)')
plt.xticks(range(0, 24))
plt.grid(True, alpha=0.2)
plt.legend(title='Akut sag?')

plt.show()
```

```python
import plotly.express as px


# grupperer data for at få gennemsnittet per time for hvert speciale
df_plot = df_otte_til_seksten.groupby(['Time', 'Speciale'])['Forsinkelse (minutter)'].mean().reset_index()

# 2. Opret line plottet
fig = px.line(
    df_plot, 
    x='Time', 
    y='Forsinkelse (minutter)', 
    color='Speciale',
    title='Gennemsnitlig forsinkelse per speciale (8:00 - 16:00)',
    labels={'Time': 'Klokkeslæt', 'Forsinkelse (minutter)': 'Gennemsnitlig forsinkelse (min)'},
    markers=True  # Tilføjer prikker på linjen for hver time
)

#Start med at alle er skjulte
fig.update_traces(visible="legendonly")


# 5. Juster layout (24h format på x-aksen)
fig.update_layout(
    xaxis=dict(
        tickmode='linear', 
        tick0=8, 
        dtick=1,
        range=[7.5, 16.5] #  valgte timer
    ),
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=1.02
    ),
    margin=dict(r=150),
    hovermode="x unified"
)

fig.show()
```
